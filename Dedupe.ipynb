```python
df2 = dedupe_new[dedupe_new['ID'].duplicated() == True]
```

```python
after_dedupe = dedupe_new.copy(deep=True)
df = after_dedupe[after_dedupe['confidence'].notnull()]
df = df.reset_index(drop=True)
```

```python
origin_list = df_origin['ID'].to_list()
mian_record_after_dedupe = df.loc[df['ID'].isin(origin_list)]
mian_record_after_dedupe = mian_record_after_dedupe.reset_index(drop=True).sort_values('cluster id')

duplicate_list = df_duplicate['ID'].to_list()
duplicate_record_after_dedupe = df.loc[df['ID'].isin(duplicate_list)]
duplicate_record_after_dedupe = duplicate_record_after_dedupe.reset_index(drop=True).sort_values('cluster id')
```

```python
mian_record_after_dedupe
```

```python
duplicate_record_after_dedupe
```

```python
df_id = mian_record_after_dedupe['cluster id']
df_confidence = mian_record_after_dedupe['confidence']
```

```python
mian_record_after_dedupe.drop(['cluster id','confidence'], axis=1, inplace=True)
duplicate_record_after_dedupe.drop(['cluster id','confidence'], axis=1, inplace=True)
duplicate_record_after_dedupe.rename(columns={'Citizenship': 'Matched Citizenship',
                                              'Date of Birth': 'Matched Date of Birth' ,
                                              'Ethnicity': 'Matched Ethnicity',
                                              'First Name': 'Matched First Name',
                                              'ID': 'Matched ID',
                                              'Indigenous Identity': 'Matched Indigenous Identity',
                                              'Language (Mother Tongue)': 'Matched Language (Mother Tongue)',
                                              'Last Name': 'Matched Last Name',
                                              'Mother\'s Birth Last Name': 'Matched Mother\'s Birth Last Name',
                                              'Place of Birth': 'Matched Place of Birth',
                                              'Race': 'Matched Race',
                                              'Religion': 'Matched Religion'}, inplace=True)
```

```python
final_df = pd.concat([mian_record_after_dedupe,duplicate_record_after_dedupe,df_confidence,df_id],axis=1)
```
