```python
# import libraries
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
# Load the main record
main = pd.read_excel('Main Records.xlsx')
main.head()
```

```python
# Load the Duplicates
Duplicates = pd.read_excel('Duplicates.xlsx')
Duplicates.head()
```

```python
# Merge two tables
merge = pd.concat([main,Duplicates])
merge.head()
merge.info
```

```python
merge["Mother's Birth Last Name"].dropna(axis=0, inplace=False, how=None)
```

```python
# Combine all columns
merge["combine"]=merge["Last Name"].astype(str)+' '+merge["First Name"].astype(str)+' '+merge["Mother's Birth Last Name"].astype(str)+' '+merge["Indigenous Identity"].astype(str)+' '+merge["Religion"].astype(str)+' '+merge["Ethnicity"].astype(str)+' '+merge["Race"].astype(str)+' '+merge["Place of Birth"].astype(str)+' '+merge["Citizenship"].astype(str)+' '+merge["Language (Mother Tongue)"].astype(str)
```

```python
merge["combine"].info
```

```python
merge["combine"].head()
```

```python
documents = merge["combine"].values.astype("U")
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents)
```

```python
k = 10000
model = KMeans(n_clusters=k, init='k-means++', max_iter=100, n_init=1)
model.fit(features)
```

```python
merge['cluster'] = model.labels_
```

```python
merge.head()
```

```python
print("Cluster centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()

for i in range(k):
    print("Cluster %d:" % i)
    for j in order_centroids[i, :10]: #print out 10 feature terms of each cluster
        print (' %s' % terms[j])
    print('------------')
```

```python
# output the result to a text file.

clusters = merge.groupby('cluster')    

for cluster in clusters.groups:
    f = open('cluster'+str(cluster)+ '.csv', 'w') # create csv file
    data = clusters.get_group(cluster)[['First Name','Last Name']] # get title and overview columns
    f.write(data.to_csv(index_label='ID')) # set index to id
    f.close()
```
